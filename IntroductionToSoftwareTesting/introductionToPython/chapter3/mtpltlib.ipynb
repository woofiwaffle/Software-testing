```python
import matplotlib.pyplot as plt
import numpy as np


def linear_plot():
    x = [1, 5, 10, 15, 20]
    y1 = [4, 3, 1, 8, 22]
    y2 = [4, 3, 1, 8, 12]
    plt.figure(figsize=(12, 7))
    plt.plot(x, y1)
    plt.plot(x, y2)
    plt.legend()
    plt.grid()
    plt.show()


def area_plot():
    plt.fill_between([-1, 0, 1], [2.3, 1.9, 2.2])
    plt.show()


def hist_plot():
    data = np.random.randint(0, 100, 100)
    plt.hist(data, bins=20)
    plt.show()


def bar_plot():
    index = np.arange(5)
    values = [5, 7, 3, 4, 6]
    plt.bar(index, values)
    plt.xticks(index + 0.4, ['A', 'B', 'C', 'D', 'E'])
    plt.show()


def pie_plot():
    values = [24, 17, 53, 21, 35]
    labels = ['Ford', 'Toyota', 'BMV', 'AUDI', 'Jaguar']
    plt.pie(values, labels=labels)
    plt.show()


def box_plot():
    data = [24, 17, 53, 21, 35]
    plt.boxplot(data)
    plt.show()


linear_plot()
area_plot()
hist_plot()
bar_plot()
pie_plot()
box_plot()
```
